```python
from __future__ import print_function
from ipywidgets import interact, interactive, fixed, interact_manual
import ipywidgets as widgets

import numpy as np
imp
import plotly.express as px
```

```python
def plot(data,x='x',y='y'):
    fig = px.line(y=data,labels={'x':x,'y':y})
    fig.show()
def load(file):
    return np.loadtxt(open(file, "r"), delimiter=",", skiprows=0)
```

```python
ref = load('Person_C/Reference.csv')
mean_ref = np.mean(ref)
print(mean_ref)
plot(ref,'t','mV')
```

```python
raw = load('Person_C/Wrist_flex2.csv')
```

```python
#1 rectify
data = raw-mean_ref
rec = np.absolute(data)
plot(rec)
```

```python
#2 MA - not working
window_size = 100
window = np.ones(window_size) / window_size
ma_data_convolved = np.convolve(rec, window, mode='valid')
plot(ma_data_convolved)
```

```python
#3 RMS
rms_sig = np.sqrt(np.multiply(rec,ref))
window_size = 100
window = np.ones(window_size) / window_size
rms_data_convolved = np.convolve(rms_sig, window, mode='valid')
plot(rms_data_convolved)
```

```python
import matplotlib.pyplot as plt
import scipy.fftpack

# Number of samplepoints
N = 600
# sample spacing
T = 1.0 / 800.0
x = np.linspace(0.0, N*T, N)
y = np.sin(50.0 * 2.0*np.pi*x) + 0.5*np.sin(80.0 * 2.0*np.pi*x)
yf = scipy.fftpack.fft(y)
xf = np.linspace(0.0, 1.0/(2.0*T), N//2)

fig, ax = plt.subplots()
ax.plot(xf, 2.0/N * np.abs(yf[:N//2]))
plt.show()
```
